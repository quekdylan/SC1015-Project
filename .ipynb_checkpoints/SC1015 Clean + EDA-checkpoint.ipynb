

## Background:

In the dynamic automotive industry, accurately estimating vehicle prices is crucial for car owners looking to buy or sell their vehicles. However, crucial factors are often overlooked by traditional methods of pricing vehicles and leading to inaccurate price estimations.

## Objective:

The objective is to help consumers analyse trends in the automotive market to estimate vehicle prices based on various attributes. By leveraging a dataset containing individual vehicle sales transactions and associated attributes, including condition, mileage, brand, and more, we aim to provide valuable insights and tools for market analysis and price prediction.

## 1. Data Preparation


### 1.1 Identify Issues/Irregularities
Check for NULL values and other irregularities

```python
# Check for NULL values
sales_data.isnull().sum()
```

```python
# Checking for other types of empty/unknown values (eg. "—")
for column in sales_data.columns:
    unique_values = sales_data[column].unique()
    print(f"Unique values in column '{column}': {unique_values}\n")
```

### 1.2 Data preparation / cleaning
1. Rename column names to more understandable names
2. Drop rows with NULL values or irregularities (from part 1)
3. Change sale date to a more standardized format

```python
# Rename of column names to new understandable names
sales_data.rename(columns={
    "year": "Manufacturing Year",
    "make": "Brand",
    "trim": "Model Version",
    "body": "Vehicle Type",
    "transmission" : "Gear",
    "vin" : "Vehicle ID",
    "state" : "Registration State",
    "odometer" : "Mileage",
    "color" : "Exterior Colour",
    "interior" : "Interior Colour",
    "mmr" : "Market Value",
    "sellingprice" : "Selling Price",
    "saledate" : "Sale Date"
}, inplace=True)

# Convert all variable names to uppercase
cleaned_data = sales_data
cleaned_data.columns = cleaned_data.columns.str.upper()
cleaned_data.head()
```

```python
# Drop missing values
cleaned_data.dropna(inplace=True)
cleaned_data.isnull().sum()
```

```python
# Drop rows with '—' values in interior colour and exterior colour 
cleaned_data = cleaned_data[cleaned_data['INTERIOR COLOUR'] != '—']
cleaned_data = cleaned_data[cleaned_data['EXTERIOR COLOUR'] != '—']

for column in cleaned_data.columns:
    unique_values = cleaned_data[column].unique()
    print(f"Unique values in column '{column}': {unique_values}\n")
```

```python
# Convert the date column to datetime format
cleaned_data["SALE DATE"] = pd.to_datetime(cleaned_data["SALE DATE"], utc=True)

# Format the date column as "DD/MM/YYYY"
cleaned_data["SALE DATE"] = cleaned_data["SALE DATE"].dt.strftime("%d/%m/%Y")

# Print the updated DataFrame
cleaned_data.head()
```

```python
# Ensure there is still enough rows after dropping
cleaned_data.info()
```

#### 2.2 Univariate Analysis
1. Separate Numerical and Categorical data
2. Plot distributions of numerical data
3. Plot distributions of categorical data


```python
# Separate Numerical and Categorical Variables
numerical_columns = cleaned_data.select_dtypes(include=['int64','float64']).columns
categorical_columns = cleaned_data.select_dtypes(include=['object']).columns

print("Numerical Columns\n", numerical_columns.values)
print("\nCategorical_Columns\n", categorical_columns.values)
```

```python
# Distributions of numerical variables

for col in numerical_columns:
    f, axes = plt.subplots(1, 3, figsize=(24, 6))
    sb.boxplot(data = cleaned_data[col], orient = "h", ax = axes[0])
    sb.histplot(data = cleaned_data[col], ax = axes[1])
    sb.violinplot(data = cleaned_data[col], orient = "h", ax = axes[2])
    plt.show()
```

```python
# Distributions of categorical variables

# Find number of unique values for categorical columns
unique_values_count = {column: cleaned_data[column].nunique() for column in categorical_columns}

# Plot distributions for columns with less than 100 unique values
for key, value in unique_values_count.items():
    if value < 100:
        plt.figure(figsize=(12, 18))
    
        # Skip SALE DATE
        if key == 'SALE DATE':
            continue
            
        sb.histplot(y=cleaned_data[key])
        # Add a title and show the plot
        plt.title(f'{key} Distribution')
        plt.show()
```

```python
# Load the map of the United States
states_url = 'https://www2.census.gov/geo/tiger/GENZ2021/shp/cb_2021_us_state_20m.zip'
usa = gpd.read_file(states_url)

# Ensure that the state codes are in uppercase
df = pd.DataFrame()
df['State'] = cleaned_data['REGISTRATION STATE'].str.upper()

# Calculate state frequencies
state_counts = df['State'].value_counts().reset_index()
state_counts.columns = ['STUSPS', 'Frequency']  # Make sure column names match those in the shapefile

# Merge the frequency data with the geodataframe
usa = usa.merge(state_counts, on='STUSPS', how='left')

# Replace NaN with 0 in the 'Frequency' column for states with no data
usa['Frequency'].fillna(0, inplace=True)

# Plotting
fig, ax = plt.subplots(1, figsize=(20, 10))
usa.plot(column='Frequency', cmap='YlGnBu', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True,
         legend_kwds={'label': "Registration Frequency by State", 'orientation': "horizontal"})
plt.title('Registration Frequency by U.S. State')
plt.show()
```

#### 2.3 Bivariate Analysis
1. Plot numeric predictors against response (Sale Price)
2. Correlation Matrix between numeric variables
3. Plot categorical predictors against response (Sale Price)

```python
# Correlation between numerical variables and sale price
for col in numerical_columns:
    #f, axes = plt.subplots(10, 2, figsize=(24, 6))
    # Create a joint dataframe by concatenating the two variables
    df = pd.concat([cleaned_data[col], cleaned_data["SELLING PRICE"]], axis = 1).reindex(cleaned_data[col].index)
    
    # Jointplot    
    sb.jointplot(data = cleaned_data, x = col, y = "SELLING PRICE", height = 12)
    plt.show()
```

```python
# Correlation Matrix between numeric variables
numeric_df = cleaned_data.select_dtypes(include=['int64','float64'])
f = plt.figure(figsize=(12, 8))
sb.heatmap(numeric_df.corr(), vmin = -1, vmax = 1, annot = True, fmt = ".2f")
plt.title('Correlation Matrix between numeric variables', fontsize=20)
plt.show()

sb.pairplot(data = numeric_df)
plt.show()
```

```python
# Categorical variables against sale price

# Plot distributions for columns with less than 100 unique values
for col, value in unique_values_count.items():
    if value < 100:
        # Create a joint dataframe by concatenating the two variables
        df = pd.concat([cleaned_data[col], cleaned_data["SELLING PRICE"]], axis = 1)
        # Jointplot    
        f = plt.figure(figsize=(12, 18))
        sb.boxplot(x = "SELLING PRICE", y = col, data = df)
        plt.show()
```
